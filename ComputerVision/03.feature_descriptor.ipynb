# Local Binary Pattern (LBP)

```python
from skimage.feature import local_binary_pattern
import cv2
import matplotlib.pyplot as plt


# settings for LBP
METHOD = 'uniform'
radius = 3
n_points = 8 * radius

img = cv2.imread('./picture/son.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img,'gray')
plt.show()

lbp = local_binary_pattern(img, n_points, radius, METHOD)
plt.imshow(lbp,'gray')
plt.show()
```

# Histogram of oriented gradient (HOG)

```python
import matplotlib.pyplot as plt

from skimage.feature import hog
from skimage import data, exposure

#    
# https://scikit-image.org/docs/dev/api/skimage.data.html
image = data.astronaut()
fd, hog_image = hog(image,orientations=9, pixels_per_cell=(8,8), cells_per_block=(1,1),visualize=True, multichannel=True)
fig,(ax1,ax2)= plt.subplots(1,2,figsize=(10,8),sharex=True,sharey=True)

ax1.axis('off')
ax1.imshow(image,cmap=plt.cm.gray)
ax1.set_title('Input image')

hog_image_rescaled = exposure.rescale_intensity(hog_image,in_range=(0,100))

ax2.axis('off')
ax2.imshow(hog_image_rescaled, cmap=plt.cm.gray)
ax2.set_title('HOG')
plt.show()
```
